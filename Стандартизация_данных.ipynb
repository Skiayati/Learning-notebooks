2. В исходных данных найдите потенциальные выбросы по признаку price_doc с помощью метода z-отклонения. Используйте логарифмический масштаб распределения. Сделайте «послабление» на 0.7 сигм в обе стороны распределения. Сколько выбросов вы получили?

```python
def outliers_z_score(data, feature, log_scale=False, left=3, right=3):
  if log_scale:
    x = np.log(data[feature] + 1)
  else:
    x = data[feature]
  mu = x.mean()
  sigma = x.std()
  lower_bound = mu - left * sigma
  upper_bound = mu + left * sigma
  outliers = data[(x < lower_bound) | (x > upper_bound)]
  return outliers

outliers = outliers_z_score(df, 'price_doc', True)
print(outliers.shape[0]) # 146 потенциальных выбросов

outliers = outliers_z_score(df, 'price_doc', True, 3.7, 3.7) # «послабление» на 0.7 сигм в обе стороны распределения
print(outliers.shape[0]) # 52 потенциальных выброса
```

3. Добавьте фишку с логарифмированием в свою функцию outliers_iqr(). Добавьте в неё параметр log_scale. Если он выставлен в True, то производится логарифмирование признака. Примените полученную функцию к признаку price_doc. Число межквартильных размахов в обе стороны обозначьте как 3.

```python
def outliers_iqr(data, feature, log_scale=False, left=1.5, right=1.5):
  if log_scale:
    x = np.log(data[feature] + 1)
  else:
    x = data[feature]
  quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75)
  iqr = quartile_3 - quartile_1
  lower_bound = quartile_1 - (iqr * left)
  upper_bound = quartile_3 + (iqr * right)
  outliers = data[(x < lower_bound) | (x > upper_bound)]
  return outliers

outliers = outliers_iqr(df, 'price_doc', False, 3, 3) # Без логарифмирования
print(outliers.shape[0]) # 629 потенциальных выброса

outliers = outliers_iqr(df, 'price_doc', True, 3, 3) # С логарифмированием
print(outliers.shape[0]) # 92 потенциальных выброса
```
